```python
import math

radio1 = 3 
radio2 = 8 
radio3 = 10 

circunferencia1 = 2 * math.pi * radio1
circunferencia2 = 2 * math.pi * radio2
circunferencia3 = 2 * math.pi * radio3

print("la circunferencia para un radio de", radio1, "es", circunferencia1)
print("la circunferencia para un radio de", radio2, "es", circunferencia2)
print("la circunferencia para un radio de", radio3, "es", circunferencia3)
```
